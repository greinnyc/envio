```python
from datetime import datetime, timedelta
import random
import math
class MYExtractFinal(Extract):
    
    def __init__(self, data, since='2019-01-21', until='2019-01-27', until_train='2019-01-20'):
        self.since = since
        self.until = until
        self.until_train = until_train
        self.data = data.copy()
        initial_size = data.shape[0]
        
        self.list_vigentes = self.get_vigentes()
        print(data.shape, data[data['CodigoSapCanilla'].isin(self.list_vigentes)].shape)
        
        self.data = data[data['CodigoSapCanilla'].isin(self.list_vigentes)]
        
        self.no_vigentes = random.sample(list(data[~data['CodigoSapCanilla'].isin(self.list_vigentes)]['CodigoSapCanilla'].unique()),k=15)
        
        fig, axes = plt.subplots(nrows=len(self.no_vigentes), ncols=1, figsize=(15, 5*len(self.no_vigentes)))
        for ix, codigo in enumerate(self.no_vigentes):
            data[data['CodigoSapCanilla'] == codigo]['VentaNeta'].plot(legend=True, title=str(codigo), ax=axes[ix])
        
    def get_vigentes(self):
        hace_3_meses = (datetime.now() - timedelta(days=90)).strftime('%Y-%m-%d')
        print('hace_3_meses :: ', hace_3_meses)
        return list(self.data[self.data['Pauta'] > 0 ][hace_3_meses:]['CodigoSapCanilla'].unique())
        
        
    def get_last_days(self, since):
        ###### Agrupando por codigo de canilla y fecha al rango de datos a predecir
        last_date_canilla = self.data[since:].groupby(
            by=['CodigoSapCanilla', 'Fecha']
        )[[col_venta]].max()
        
        ###### AÃ±adiendo Codigo canilla al agrupado
        last_date_canilla['index'] = last_date_canilla.index
        last_date_canilla['CodigoSapCanilla'] = last_date_canilla['index'].apply(
            lambda _: _[0]
        )
        return last_date_canilla
    
    @staticmethod
    def get_with_null(last_date_canilla):
        rows_with_null = last_date_canilla[
            (last_date_canilla[col_venta].isnull()) | \
            (str(last_date_canilla[col_venta]).lower() == 'nan')
        ]
        return list(rows_with_null['CodigoSapCanilla'].unique())
    
    def procces_day(self, data, code):
        df3 = data.copy()
        col, new_col,date = 'VentaNeta', 'new','Fecha'
        #dat=
        
        """
        METODO DE AUTOCOMPLETADO 
        """
        print(code, df3.shape)
        if df3[col][-4:].mean() == df3[col][-1] and df3[col][-4:].mean() == df3[col][-2] and df3[col][-4:].mean() < 2:
            print("Se detecto una relacion lineal : ", code)
            return df3[col]
        
        #fecha=
        dat=df3[np.isnan(df3[col])][date]
        if not (dat is None):
            mean = df3[col].median()
        else:
            mean = df3[df3[date]< dat][col].mean()
            
            
              
        # df3[df3[Fecha]< (df3[np.isnan(df3[col])][Fecha])][col].median()
        
        df3[col] = df3[['Pauta', col]].apply(
            lambda _: _[col] if _['Pauta'] > 0 else np.nan, axis=1
        )
        df3[col].fillna(mean, inplace=True)
        #print("se modifico dato : ", str(float(mean)),code)
        return df3[col] 
        
    def get_data_pauta_weekday(self, data_day, code):
        # Obtenemos las columnas de VEntaNeta y Pauta asociadas
        data_canilla = data_day[
            data_day['CodigoSapCanilla'] == code
        ][['VentaNeta', 'Pauta', 'Fecha', 'weekday']][:self.until_train].copy()
        
        # Deteccion de casos con demasiadas ausencias
        if data_canilla[data_canilla['Pauta'] == 0].shape[0] / data_canilla.shape[0] > 0.5:
            return 'Cantidad de ausencia supera el 50% de casos reportados para este dia'
        
        return self.procces_day(data_canilla, code)
    
    def populate_data(self, code_day):
        data_day = self.data[self.data['weekday'] == code_day]
        self.lista_canillas = set(
            data_day['CodigoSapCanilla'].unique()
        )
        print('Total canillas para este dia: ', len(self.lista_canillas))
        
        dicc = {}
        for canilla in self.lista_canillas:
            response = self.get_data_pauta_weekday(data_day, canilla)
            
            if isinstance(response, str):
                print("\n CASO INVALIDO ", canilla, response)
        #        fig, axes = plt.subplots(nrows=1, ncols=1)
             #   data_day[data_day['CodigoSapCanilla'] == canilla]['VentaNeta'].plot(
              #      legend=True, ax=axes, title=canilla
               # )
            else:
                dicc[canilla] = response
        return dicc
    
    @staticmethod
    def valid_null(dicc_data):
        cant_nulos = 0
        
        for k, v in dicc_data.items():
            for num_day, datus in v.items():
                nulos = datus.isnull().sum()
                if nulos:
                    print(k, ' __ ', num_day ,' : ', nulos, 'de', v.get('data').shape[0])
                    cant_nulos +=1
        
        print('\n {} nulos detectados'.format(cant_nulos))
```

```python
def noise_control(dicc_data):
    for k, v in dicc_data.items():
        for num_day, datus in v.items():
            #fig, axes = plt.subplots(nrows=1, ncols=1)
            #datus.plot(legend=True, ax=axes, title='{} - {}'.format(k, num_day))
            
            q75, q25 = np.percentile(datus, [75 ,25])
            iqr = q75 - q25
            size_outlier = (datus.max() - datus.min()) - iqr
            
            if size_outlier < 5:
                cut_up = 97 / 100
                cut_down = 3 / 100
            elif size_outlier < 10:
                cut_up = 95 / 100
                cut_down = 5 / 100
            else:
                cut_up = 90 / 100
                cut_down = 10 / 100

            quantile_max = int(datus.quantile(cut_up))
            quantile_min = int(datus.quantile(cut_down))
            
            if datus[-4:].mean() == datus[-1] and datus[-4:].mean() == datus[-2] and datus[-4:].mean() < 2:
                print("Relacion Lineal : ", k, num_day, datus[-5:].values, datus[-1])
            else:
                print("()() ", k , cut_up, '=', quantile_max, '/// ', cut_down, '=', quantile_min)
                v[num_day] = datus.apply(lambda x: reemplace_percentile(x, quantile_min, quantile_max))
            # v[num_day].plot(legend=True, ax=axes)
            
    return dicc_data
```

```python
def reemplace_percentile(val, val_smaller, val_bigger):     
    if val < val_smaller:
        return val_smaller
    elif val > val_bigger:
        return val_bigger
    return val
```
